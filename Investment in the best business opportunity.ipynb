```python
import urllib.request
# -----------------------------------------------------------------------------
# Name:        html_table_parser
# Purpose:     Simple class for parsing an (x)html string to extract tables.
#              Written in python3
#
# Author:      Josua Schmid
#
# Created:     05.03.2014
# Copyright:   (c) Josua Schmid 2014
# Licence:     AGPLv3
# -----------------------------------------------------------------------------

from html.parser import HTMLParser
class HTMLTableParser(HTMLParser):
    """ This class serves as a html table parser. It is able to parse multiple
    tables which you feed in. You can access the result per .tables field.
    """
    def __init__(
        self,
        decode_html_entities=False,
        data_separator=' ',
    ):

        HTMLParser.__init__(self, convert_charrefs=decode_html_entities)

        self._data_separator = data_separator

        self._in_td = False
        self._in_th = False
        self._current_table = []
        self._current_row = []
        self._current_cell = []
        self.tables = []

    def handle_starttag(self, tag, attrs):
        """ We need to remember the opening point for the content of interest.
        The other tags (<table>, <tr>) are only handled at the closing point.
        """
        if tag == 'td':
            self._in_td = True
        if tag == 'th':
            self._in_th = True

    def handle_data(self, data):
        """ This is where we save content to a cell """
        if self._in_td or self._in_th:
            self._current_cell.append(data.strip())
    
    def handle_endtag(self, tag):
        """ Here we exit the tags. If the closing tag is </tr>, we know that we
        can save our currently parsed cells to the current table as a row and
        prepare for a new row. If the closing tag is </table>, we save the
        current table and prepare for a new one.
        """
        if tag == 'td':
            self._in_td = False
        elif tag == 'th':
            self._in_th = False

        if tag in ['td', 'th']:
            final_cell = self._data_separator.join(self._current_cell).strip()
            self._current_row.append(final_cell)
            self._current_cell = []
        elif tag == 'tr':
            self._current_table.append(self._current_row)
            self._current_row = []
        elif tag == 'table':
            self.tables.append(self._current_table)
            self._current_table = []
```

Lets define a function to get a nerby venues around the mentioned districts in Lima Downtown

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            RADIUS, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Districts', 
                  'District Latitude', 
                  'District Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
num_top_venues = 5

for hood in Limadt_grouped['Districts']:
    print("----"+hood+"----")
    temp = Limadt_grouped[Limadt_grouped['Districts'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

<font size=3>
Now, let's write a function to sort the venues in descending order.
</font>

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

<font size=3>
Lets create a dataframe to display the top 10 commun categories in each districts
</font>

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Districts']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
district_venues_sorted = pd.DataFrame(columns=columns)
district_venues_sorted['Districts'] = Limadt_grouped['Districts']

for ind in np.arange(Limadt_grouped.shape[0]):
    district_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Limadt_grouped.iloc[ind, :], num_top_venues)

district_venues_sorted
```

## Clustering

<font size=3>Run k-means to cluster the neighborhood into 5 clusters.</font>

```python
# set number of clusters
kclusters = 5

limadt_grouped_clustering = Limadt_grouped.drop('Districts', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(limadt_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

<font size=3>Add column of the number clusters.</font>

```python
#drop the column if this exist
if 'Cluster Labels' in district_venues_sorted.columns:
    district_venues_sorted.drop(columns='Cluster Labels', inplace=True)
# add clustering labels
district_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Limadt_merged = Lima_dt

# merge Limadt_merged with district_venues_sorted to add latitude/longitude for each district
Limadt_merged = Limadt_merged.join(district_venues_sorted.set_index('Districts'), on='Districts')

Limadt_merged.head()
```

<font size=3>And then make a map to show the clusters points</font>

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Limadt_merged['Latitude'], Limadt_merged['Longitude'], Limadt_merged['Districts'], Limadt_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

<font size=3>
We are going to create a function that returns, in the case of the origin and destination districts, the address and, in the case of the waypoints, their coordinates. This will be done since thanks to the google maps API, the requests necessary to obtain the route that conforms to the requirements need this data.
</font>

```python
from urllib.parse import urlencode
#define waypoints (points where you want the route to pass)
def ref_point(dist,**kwarg):#**kwarg to able if is origin or destination
    #finding the index of the district
    l=list(Limadt_merged['Districts']==dist)
    n=l.index(True)

    api_key='AIzaSyAvUy32NQyef0pt0jVYSx6mZBSzG09GWzM' #Google map API key 
    #getting the latitude and longitude
    lat=Limadt_merged.loc[n,'Latitude']
    lng=Limadt_merged.loc[n,'Longitude']
    
    #We look for a place close to the coordinates found previously
    url='https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
    params={
        'key':api_key,
        'input':Limadt_merged.loc[n,'1st Most Common Venue'],
        'inputtype':'textquery',
        'fields':'formatted_address,name,geometry'
    }
    locationbias=f'point:{lat},{lng}'
    params['locationbias']=locationbias

    params_encoder=urlencode(params)
    place=f'{url}?{params_encoder}'
    r=requests.get(place)
    way_point_detail=r.json()['candidates'][0]['geometry']['location']
    latwp=way_point_detail.get('lat')
    lngwp=way_point_detail.get('lng')
    
    for key,value in kwarg.items():
        if key=='origin'or key=='destination' and value==True:
            address=r.json()['candidates'][0]['formatted_address']
            return address
        elif key=='waypoint' and value==True:
            way_point_detail=r.json()['candidates'][0]['geometry']['location']
            latwp=way_point_detail.get('lat')
            lngwp=way_point_detail.get('lng')
            return latwp,lngwp
            
    return latwp,lngwp
```

We are going to create a function that returns the json file of recommended route by google maps having as parameter the origin, destination and waypoints.
    
</font>

```python
#function that returns the route from the origin district to the destination district passing through specific points
def Get_route(origin,destination,*arg):
    api_key='AIzaSyAvUy32NQyef0pt0jVYSx6mZBSzG09GWzM'
    end_point=f'https://maps.googleapis.com/maps/api/directions/json?origin={origin},au&destination={destination},au&waypoints='
    key=f'&key={api_key}'
    url=f'https://maps.googleapis.com/maps/api/directions/json?origin={origin}&destination={destination}&key={api_key}'
    r=requests.get(url)
    route=r.json()
    i=1
    for n in arg:
        if i==1:
            param=f'via:{n[0]}%2C{n[1]}' #In case would pass only one waypoint
            url=end_point+param+key
            r=requests.get(url)
            route=r.json()
        else:
            param=param+'%7C'+f'via:{n[0]}%2C{n[1]}' #In case would have many waypoints
            url=end_point+param+key
            r=requests.get(url)
            route=r.json()
        i+=1
    return route
```
